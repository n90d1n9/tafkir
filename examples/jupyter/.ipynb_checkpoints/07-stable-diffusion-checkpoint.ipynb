# 🎨 Stable Diffusion Image Generation with Tafkir

This notebook demonstrates text-to-image generation using Tafkir's ONNX-based Stable Diffusion implementation.

## What You'll Learn
- Generate images from text prompts
- Control generation with seed, steps, and guidance scale
- Display generated images inline
- Create multiple variations

## Prerequisites
- Java 25 kernel installed
- ONNX Runtime available
- Stable Diffusion model downloaded (auto-downloads on first run)

```java
// Import required packages
import tech.kayys.tafkir.sdk.core.TafkirSdk;
import tech.kayys.tafkir.spi.inference.InferenceRequest;
import tech.kayys.tafkir.spi.inference.StreamingInferenceChunk;
import tech.kayys.tafkir.spi.model.ModalityType;

import java.nio.file.Files;
import java.nio.file.Path;
import java.util.Base64;
import java.util.concurrent.CountDownLatch;
import java.util.concurrent.TimeUnit;
import javax.imageio.ImageIO;
import java.awt.image.BufferedImage;
import java.io.ByteArrayInputStream;

System.out.println("✅ Imports successful");
```

## 1. Initialize the SDK

```java
// Initialize Tafkir SDK
TafkirSdk sdk = TafkirSdk.builder().local().build();
System.out.println("✅ SDK initialized successfully");
```

## 2. Generate Your First Image

Let's start with a simple prompt and see what we can create!

```java
// Configure generation parameters
String prompt = "a cat playing ball, photorealistic, sunny garden";
long seed = 42L;
int steps = 20;
float guidanceScale = 7.5f;
String outputFile = "cat-playing.png";

System.out.println("Prompt: " + prompt);
System.out.println("Seed: " + seed);
System.out.println("Steps: " + steps);
System.out.println("Guidance: " + guidanceScale);
System.out.println();

// Build request
InferenceRequest request = InferenceRequest.builder()
    .model("CompVis/stable-diffusion-v1-4")
    .prompt(prompt)
    .parameter("seed", seed)
    .parameter("steps", steps)
    .parameter("guidance_scale", guidanceScale)
    .streaming(true)
    .build();

// Execute with streaming
CountDownLatch latch = new CountDownLatch(1);
long startTime = System.currentTimeMillis();

sdk.streamCompletion(request)
    .subscribe().with(
        chunk -> {
            if (chunk.delta() != null && chunk.delta().startsWith("[")) {
                System.out.print("\r" + chunk.delta() + "  ");
                System.out.flush();
            }
            
            if (chunk.modality() == ModalityType.IMAGE && chunk.imageDeltaBase64() != null) {
                try {
                    byte[] pngData = Base64.getDecoder().decode(chunk.imageDeltaBase64());
                    Files.write(Path.of(outputFile), pngData);
                    
                    long duration = System.currentTimeMillis() - startTime;
                    System.out.println();
                    System.out.println("✅ Image saved: " + outputFile);
                    System.out.printf("⏱️  Time: %.1f seconds%n", duration / 1000.0);
                    System.out.printf("📊 Size: %,d bytes%n", pngData.length);
                } catch (Exception e) {
                    System.err.println("\n❌ Failed: " + e.getMessage());
                }
            }
        },
        error -> {
            System.err.println("\n❌ Error: " + error.getMessage());
            latch.countDown();
        },
        latch::countDown
    );

latch.await(600, TimeUnit.SECONDS);
```

## 3. Display the Generated Image

```java
// Load and display the image
BufferedImage img = ImageIO.read(new java.io.File(outputFile));
System.out.println("Image dimensions: " + img.getWidth() + "x" + img.getHeight());

// Display in notebook (if using Jupyter with image support)
byte[] imageBytes = Files.readAllBytes(Path.of(outputFile));
String base64 = Base64.getEncoder().encodeToString(imageBytes);
System.out.println("![Generated Image](data:image/png;base64," + base64.substring(0, 100) + "...)");

// Alternative: Use HTML img tag
System.out.println("<img src='data:image/png;base64," + base64 + "' width='512' height='512' />");
```

## 4. Generate Multiple Variations

Let's create 4 different variations by changing the seed!

```java
// Generate multiple variations
String prompt2 = "a cyberpunk cat with neon lights, digital art";
long[] seeds = {100L, 200L, 300L, 400L};
int steps2 = 15; // Fewer steps for faster generation

System.out.println("Generating " + seeds.length + " variations...");
System.out.println("Prompt: " + prompt2);
System.out.println();

for (int i = 0; i < seeds.length; i++) {
    final long seed = seeds[i];
    String variationFile = String.format("variation-seed-%d.png", seed);
    
    System.out.printf("[%d/%d] Seed %d... ", i + 1, seeds.length, seed);
    
    InferenceRequest req = InferenceRequest.builder()
        .model("CompVis/stable-diffusion-v1-4")
        .prompt(prompt2)
        .parameter("seed", seed)
        .parameter("steps", steps2)
        .parameter("guidance_scale", 7.5f)
        .streaming(true)
        .build();
    
    CountDownLatch l = new CountDownLatch(1);
    
    sdk.streamCompletion(req)
        .subscribe().with(
            chunk -> {
                if (chunk.modality() == ModalityType.IMAGE && chunk.imageDeltaBase64() != null) {
                    try {
                        byte[] png = Base64.getDecoder().decode(chunk.imageDeltaBase64());
                        Files.write(Path.of(variationFile), png);
                        System.out.println("✅ (%,d bytes)", png.length);
                    } catch (Exception e) {
                        System.out.println("❌ " + e.getMessage());
                    }
                }
            },
            error -> System.out.println("❌ " + error.getMessage()),
            l::countDown
        );
    
    l.await(600, TimeUnit.SECONDS);
}

System.out.println("\n✅ All variations generated!");
```

## 5. Display All Variations

```java
// Display all generated variations
System.out.println("<h3>Generated Variations</h3>");
System.out.println("<div style='display: flex; flex-wrap: wrap; gap: 10px;'>");

for (long seed : seeds) {
    String variationFile = String.format("variation-seed-%d.png", seed);
    if (Files.exists(Path.of(variationFile))) {
        byte[] imageBytes = Files.readAllBytes(Path.of(variationFile));
        String base64 = Base64.getEncoder().encodeToString(imageBytes);
        System.out.println(String.format(
            "<div><img src='data:image/png;base64,%s' width='256' height='256'/><br/><small>Seed: %d</small></div>",
            base64, seed
        ));
    }
}

System.out.println("</div>");
```

## 6. Experiment with Guidance Scale

Let's see how the guidance scale (CFG) affects the output!

```java
// Test different guidance scales
String prompt3 = "a majestic dragon flying over mountains";
float[] cfgScales = {5.0f, 7.5f, 10.0f, 12.0f};

System.out.println("Testing guidance scales: ");
for (float cfg : cfgScales) {
    System.out.print(cfg + " ");
}
System.out.println();
System.out.println();

for (float cfg : cfgScales) {
    String cfgFile = String.format("cfg-%.1f.png", cfg);
    System.out.printf("CFG %.1f... ", cfg);
    
    InferenceRequest req = InferenceRequest.builder()
        .model("CompVis/stable-diffusion-v1-4")
        .prompt(prompt3)
        .parameter("seed", 42L)
        .parameter("steps", 15)
        .parameter("guidance_scale", cfg)
        .streaming(true)
        .build();
    
    CountDownLatch l = new CountDownLatch(1);
    
    sdk.streamCompletion(req)
        .subscribe().with(
            chunk -> {
                if (chunk.modality() == ModalityType.IMAGE && chunk.imageDeltaBase64() != null) {
                    try {
                        byte[] png = Base64.getDecoder().decode(chunk.imageDeltaBase64());
                        Files.write(Path.of(cfgFile), png);
                        System.out.println("✅");
                    } catch (Exception e) {
                        System.out.println("❌");
                    }
                }
            },
            error -> System.out.println("❌ " + error.getMessage()),
            l::countDown
        );
    
    l.await(600, TimeUnit.SECONDS);
}

System.out.println("\n✅ Guidance scale comparison complete!");
```
