Image generation with Stable Diffusion via the Tafkir SDK.

```java
import tech.kayys.tafkir.sdk.core.TafkirSdk;
import tech.kayys.tafkir.factory.TafkirSdkFactory;
import tech.kayys.tafkir.spi.inference.InferenceRequest;
import tech.kayys.tafkir.spi.inference.StreamingInferenceChunk;
import tech.kayys.tafkir.spi.model.ModalityType;
import java.nio.file.Files;
import java.nio.file.Path;
import java.util.Base64;
import java.util.concurrent.CountDownLatch;
import java.util.concurrent.TimeUnit;

System.out.println("🔍 Tafkir JDK 25 - Pure Java AI Platform");
System.out.println("Java: " + Runtime.version());
```

```java
import tech.kayys.tafkir.sdk.core.TafkirSdk;
import tech.kayys.tafkir.factory.TafkirSdkFactory;
import tech.kayys.tafkir.spi.inference.InferenceRequest;
import tech.kayys.tafkir.spi.inference.StreamingInferenceChunk;
import tech.kayys.tafkir.spi.model.ModalityType;
import tech.kayys.tafkir.spi.Message;
import java.nio.file.Files;
import java.nio.file.Path;
import java.util.Base64;
import java.util.concurrent.CountDownLatch;
import java.util.concurrent.TimeUnit;
import java.awt.image.BufferedImage;
import javax.imageio.ImageIO;

// Connect to a running Tafkir server (start with: tafkir serve)
TafkirSdk sdk = TafkirSdkFactory.createRemoteSdk("http://localhost:8080", "community");
System.out.println("SDK initialized");
```

```java
record SDResult(String outputFile, long durationMs, long fileSize) {}

SDResult generateImage(TafkirSdk sdk, String prompt, String outputFile, long seed, int steps, float guidance) throws Exception {
    System.out.println("Generating: " + prompt);
    InferenceRequest request = InferenceRequest.builder()
        .model("CompVis/stable-diffusion-v1-4")
        .message(Message.user(prompt))
        .parameter("seed", seed)
        .parameter("steps", steps)
        .parameter("guidance_scale", guidance)
        .streaming(true)
        .build();
    CountDownLatch latch = new CountDownLatch(1);
    long startTime = System.currentTimeMillis();
    final long[] fileSize = {0};
    sdk.streamCompletion(request)
        .subscribe().with(
            chunk -> {
                if (chunk.delta() != null && chunk.delta().startsWith("[")) {
                    System.out.print(chunk.delta());
                    System.out.flush();
                }
                if (chunk.modality() == ModalityType.IMAGE && chunk.imageDeltaBase64() != null) {
                    try {
                        byte[] png = Base64.getDecoder().decode(chunk.imageDeltaBase64());
                        Files.write(Path.of(outputFile), png);
                        fileSize[0] = png.length;
                    } catch (Exception e) { System.err.println("Save failed: " + e.getMessage()); }
                }
            },
            error -> { System.err.println("Error: " + error.getMessage()); latch.countDown(); },
            latch::countDown
        );
    latch.await(600, TimeUnit.SECONDS);
    System.out.println("Saved: " + outputFile);
    return new SDResult(outputFile, System.currentTimeMillis() - startTime, fileSize[0]);
}
```

```java
SDResult result = generateImage(sdk, "a cat playing ball in sunny garden, photorealistic", "cat-demo.png", 42L, 10, 7.5f);
System.out.printf("⏱️  Duration: %.1fs%n", result.durationMs() / 1000.0);
System.out.printf("📊 File size: %,d bytes%n", result.fileSize());
```

```java
import java.awt.image.BufferedImage;
import javax.imageio.ImageIO;

if (Files.exists(Path.of("cat-demo.png"))) {
    BufferedImage img = ImageIO.read(new java.io.File("cat-demo.png"));
    System.out.printf("✅ Dimensions: %dx%d pixels%n", img.getWidth(), img.getHeight());
} else {
    System.out.println("❌ Image not found. Run previous cell first.");
}
```

```java
System.out.println("🎨 Generating 3 variations...\n");
String prompt = "a cyberpunk cat with neon lights, digital art";
for (long seed : new long[]{100L, 200L, 300L}) {
    generateImage(sdk, prompt, "variation-" + seed + ".png", seed, 10, 7.5f);
}
System.out.println("✅ All variations generated!");
```

```java
for (long seed : new long[]{100L, 200L, 300L}) {
    String file = "variation-" + seed + ".png";
    if (Files.exists(Path.of(file))) {
        BufferedImage img = ImageIO.read(new java.io.File(file));
        System.out.printf("Seed %d: %dx%d pixels%n", seed, img.getWidth(), img.getHeight());
    }
}
```

```java
System.out.printf("%-10s | %-10s | %-10s%n", "Steps", "Time (s)", "Sec/Step");
System.out.println("-".repeat(35));
for (int steps : new int[]{5, 10, 15}) {
    SDResult r = generateImage(sdk, "a cat", "bench-" + steps + ".png", 42L, steps, 7.5f);
    System.out.printf("%-10d | %-10.1f | %-10.2f%n", steps, r.durationMs()/1000.0, r.durationMs()/1000.0/steps);
}
```
