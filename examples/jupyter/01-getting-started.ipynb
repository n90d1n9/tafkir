# 01 - Getting Started with Tafkir

This notebook demonstrates the basics of the Tafkir ML framework for Java, including:
- Tensor creation and operations
- Autograd and gradient computation
- Simple neural network forward pass
- Device selection (CPU, CUDA, Metal)

```java
// Import Tafkir ML framework
import tech.kayys.tafkir.ml.Tafkir;
import tech.kayys.tafkir.ml.tensor.Tensor;
import tech.kayys.tafkir.ml.autograd.GradTensor;
import tech.kayys.tafkir.ml.nn.Linear;
import tech.kayys.tafkir.ml.nn.ReLU;
import tech.kayys.tafkir.ml.nn.Sequential;

// Check available devices
System.out.println("CUDA available: " + Tafkir.isCudaAvailable());
System.out.println("Metal available: " + Tafkir.isMetalAvailable());
System.out.println("Default device: " + Tafkir.defaultDevice());

// Print framework info
Tafkir.printInfo();
```

## Tensor Creation

```java
// Create tensors
Tensor x = Tafkir.randn(3, 4).requiresGrad(true);
Tensor y = Tafkir.randn(4, 5);
Tensor z = Tafkir.zeros(3, 5);

System.out.println("x shape: " + java.util.Arrays.toString(x.shape()));
System.out.println("y shape: " + java.util.Arrays.toString(y.shape()));
System.out.println("z shape: " + java.util.Arrays.toString(z.shape()));
System.out.println("x requires grad: " + x.requiresGrad());
```

## Tensor Operations

```java
// Basic operations
Tensor a = Tafkir.tensor(new float[]{1.0f, 2.0f, 3.0f, 4.0f}, 2, 2);
Tensor b = Tafkir.tensor(new float[]{5.0f, 6.0f, 7.0f, 8.0f}, 2, 2);

Tensor c = a.add(b);
Tensor d = a.mul(b);

System.out.println("a + b = " + c);
System.out.println("a * b = " + d);
System.out.println("a @ b = " + a.matmul(b));
```

## Autograd

```java
// Create tensors with gradient tracking
GradTensor w = (GradTensor) Tafkir.randn(2, 2, true).requiresGrad(true);
GradTensor x_grad = (GradTensor) Tafkir.randn(2, 2, true).requiresGrad(true);

// Forward pass
GradTensor out = (GradTensor) w.matmul(x_grad).sum();

System.out.println("Forward pass output: " + out);

// Backward pass
out.backward();

System.out.println("w.grad: " + w.grad());
System.out.println("x.grad: " + x_grad.grad());
```

## Simple Neural Network

```java
// Build a simple feedforward network
Sequential model = Sequential.builder()
    .add(new Linear(784, 256))
    .add(new ReLU())
    .add(new Linear(256, 128))
    .add(new ReLU())
    .add(new Linear(128, 10))
    .build();

System.out.println("Model architecture:");
System.out.println("  Input: 784 (28x28 MNIST)");
System.out.println("  Hidden 1: 256 + ReLU");
System.out.println("  Hidden 2: 128 + ReLU");
System.out.println("  Output: 10 (digits 0-9)");

// Forward pass with random input
GradTensor batchInput = (GradTensor) Tafkir.randn(32, 784).requiresGrad(true);
GradTensor output = (GradTensor) model.forward(batchInput);

System.out.println("\nForward pass:");
System.out.println("  Input shape: [32, 784]");
System.out.println("  Output shape: " + java.util.Arrays.toString(output.shape()));
```

## No-Grad Mode

```java
// Inference mode (no gradient tracking)
try (var noGrad = Tafkir.noGrad()) {
    Tensor input = Tafkir.randn(1, 784);
    Tensor inferenceOutput = model.forward(input);
    
    System.out.println("Inference output shape: " + java.util.Arrays.toString(inferenceOutput.shape()));
    System.out.println("No grad tracking in inference mode");
}
```
